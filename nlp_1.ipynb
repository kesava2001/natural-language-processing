```python
#metamorphosis text
filename = '/content/drive/MyDrive/Colab Notebooks/NLP/data sets/metamorphosis.txt'
```

```python
import string, re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
sw = set(stopwords.words('english'))

file = open(filename, 'rt')
text = file.read()
file.close()

tokens = word_tokenize(text)
def cleaning(tokens):
  tokens = [t.lower() for t in tokens]
  tokens = [re.sub('[^a-zA-Z]', ' ', t) for t in tokens]
  tokens = [t for t in tokens if t.isalpha()]
  tokens = [t for t in tokens if t not in sw]
  return tokens

token = cleaning(tokens)

token[:10]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

# words = word_tokenize(text)
stemmed = [ps.stem(w) for w in token]

stemmed[:10]
```

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()

sample = ['the quick brown fox jumped over the lazy dog']
vectorizer.fit(sample)

vector = vectorizer.transform(sample)

print(vector.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import sent_tokenize
tfidf = TfidfVectorizer()
sents = sent_tokenize(text)
tfidf.fit(sents)

trans = tfidf.transform([sents[6]])
trans.toarray()
```

```python
import keras
from keras.preprocessing.text import text_to_word_sequence

tws = text_to_word_sequence(text)
tws[:10]
```

```python
from keras.preprocessing.text import one_hot
vocab_size = len(tws)
print(vocab_size)
oh = one_hot(text, vocab_size)
oh[:10]
```

```python
from keras.preprocessing.text import hashing_trick

ht = hashing_trick(text, vocab_size, hash_function='md5')
ht[:10]
```

```python
from keras.preprocessing.text import Tokenizer
from nltk.tokenize import sent_tokenize

sent = sent_tokenize(text)

t = Tokenizer()
t.fit_on_texts(sent)

print(t.word_counts)
print(t.document_count)
print(t.word_index)
print(t.word_docs)


encoded = t.texts_to_matrix(sent, mode='count')
encoded
```
